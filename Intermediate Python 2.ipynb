```python
# Simulate random walk n times
import numpy as np
import matplotlib.pyplot as plt


n = input('How many times do you want to simulate? ')
all_walks = []
for i in range(int(n)) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)

        # Implement clumsiness
        if np.random.rand() <= 0.001 : # there would be a small chance that the individual will fall from the tower and start over again
            step = 0

        random_walk.append(step)
    all_walks.append(random_walk)

# Create and plot np_aw_t
np_aw_t = np.transpose(np.array(all_walks)) # we transpose the array to make a better plot
plt.plot(np_aw_t)

plt.title('Simulation of Random Walks')
plt.ylabel('Distance from the origin')
plt.xlabel('Number of steps taken')
plt.show()
```

```python
# Select last row from np_aw_t: ends
ends = np_aw_t[-1,:]

# Plot histogram of ends, display plot
plt.hist(ends)
plt.title('Distribution of Random Walks')
plt.ylabel('Distance from the origin')
plt.xlabel('Number of Simulations')
plt.show()
```

```python
# What's the percentage of the walks reached greater than some arbitrary number of steps?
arb_value = input('What is the distance cut-off value? ')
pct = len( ends [ ends > float(arb_value) ])/float(n)

fig, ax = plt.subplots(figsize=(10,5))

ax.hist(ends, histtype='step');
ax.set(xlabel='Number of Simulations', ylabel='Distance from the origin', title = 'Distribution of Random Walks')
ax.axvline(float(arb_value), c='r', label=f"{pct*100:.3} %")
ax.legend()

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.tick_params(left=True, bottom=True,
               labelleft=True, labelbottom=True)

print('There is ' + str(pct*100) + "% " + "of the Random Walks that reached further than " + str(arb_value) + "steps.")
```
